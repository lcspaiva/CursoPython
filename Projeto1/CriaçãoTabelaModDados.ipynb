#### Entendendo os produtos

```python
dict_produtos = {}
produtos = tabela_vendas["Produto"]
for item in produtos:
    if dict_produtos.get(item, 0) == 0:
        dict_produtos[item] = 1
    else:
        dict_produtos[item] += 1
```

```python
dict_produtos_limpos = {}
for item in dict_produtos:
    espaco = item.find(' ')
    if espaco == -1:
        dict_produtos_limpos[item] = 1
    else:
        dict_produtos_limpos[item[:espaco]] = 1
```

```python
print((dict_produtos_limpos))
```

```python
print(dict_produtos)
```

#### Limpando caracteristicas dos produtos

```python
coluna_produto = tabela_vendas["Produto"]
coluna_limpa = []
for item in coluna_produto:
    espaco = item.find(' ')
    if espaco == -1:
        coluna_limpa.append(item)
    else:
        coluna_limpa.append(item[:espaco])
```

```python
tabela_vendas_mod = tabela_vendas
```

```python
tabela_vendas_mod["Tipo"] = coluna_limpa
```

```python
tabela_vendas_mod
```

### Relatório de vendas totais por estabelecimento

```python
tabela_vendas_shopp = tabela_vendas_mod[["ID Loja", "Valor Final"]]
tabela_vendas_shopp = tabela_vendas_shopp.groupby("ID Loja").sum()
tabela_vendas_shopp = tabela_vendas_shopp.sort_values(by="Valor Final", ascending = False)
tabela_vendas_shopp.to_excel("files/relatVendasTotais.xlsx") 
```

```python
tabela_vendas_shopp
```

### Relatório de quantidade de itens vendidos por estabelecimento

```python
tabela_vendas_quant_shopp = tabela_vendas_mod[["ID Loja", "Quantidade"]]
tabela_vendas_quant_shopp = tabela_vendas_quant_shopp.groupby("ID Loja").sum()
tabela_vendas_quant_shopp = tabela_vendas_quant_shopp.sort_values(by="Quantidade", ascending = False)
tabela_vendas_quant_shopp.to_excel("files/relatQntItens.xlsx") 
```

```python
tabela_vendas_quant_shopp
```

#### Relatório Ticket medio por estabelecimento

```python
tabela_vendas_media = (tabela_vendas_shopp['Valor Final']/tabela_vendas_quant_shopp['Quantidade']).to_frame()
tabela_vendas_media = tabela_vendas_media.rename(columns={0:"Ticket Medio"})
tabela_vendas_media = tabela_vendas_media.sort_values(by="Ticket Medio", ascending = False)
tabela_vendas_media.to_excel("files/relatTicketMedio.xlsx")
```
